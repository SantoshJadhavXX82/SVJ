# The Gradient Descent Algorithm to find the minima of a function !

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(-10,10)
print(x)
```

```python
#y=x2-2x+1
y = x**2 - 2*x + 1
print(y)
```

```python
print(x.shape)
print(y.shape)
```

```python
plt.plot(x, y)
```

```python
# minima function
d = []
def minima(f ,t):
    for i in range(len(f)-1):
        diff = ((f[i+1] - f[i]) / (t[i+1] - t[i]))
        print('Diff={} , f[i+1] - f[i] = {}, t[i+1] - t[i] = {}'.format(diff,(f[i+1] - f[i]), (t[i+1] - t[i])))
        d.append(diff)
    return d
```

```python
dydx = minima(y,x)

#print(d)
plt.plot(x,y,label='y v/s x')
plt.plot(x[:-1],dydx,label='dy/dx v/s x')
plt.legend()
```
