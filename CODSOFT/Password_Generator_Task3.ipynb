```python
import tkinter as tk
from tkinter import Label, Entry, Button, StringVar
import random
import string
```

```python
lcnf_label1 = {'font':('Arial', 12, 'bold'), 
              'fg':'black', 
              'width':25}

lcnf_grid_label1 = {'row':0, 'column':0, 'padx':5, 'pady':10, 'sticky':'w'}

lcnf_entry = {'background':'black', 
              'borderwidth':2, 
              'font':('Arial', 14), 
              'foreground':'white', 
              'width':25, 'justify':"center", 'cursor':'', 'insertofftime':0, 'insertontime':500}

lcnf_grid_entry = {'row':1, 'column':0, 'padx':5, 'pady':10, 'sticky':'w'}

lcnf_generate_button = {'row':5, 'column':0, 'padx':5, 'pady':10}

lcnf_label2 = {'font':('Arial', 12, 'bold'), 
              'fg':'black', 
              'width':25}

lcnf_grid_label2 = {'row':6, 'column':0, 'padx':5, 'pady':10, 'sticky':'w'}


lcnf_entry2 = {'background':'black', 
              'borderwidth':2, 
              'font':('Arial', 14), 
              'foreground':'#594D58', 
              'width':25,
              'state':'readonly', 'justify':"center"}

lcnf_grid_entry2 = {'row':9, 'column':0, 'padx':5, 'pady':10, 'sticky':'w'}
```

```python
class PasswordGeneratorApp:
    def __init__(self, master):
        self.master = master
        self.master.title("Random Password Generator")

        self.password_var = StringVar()

        self.length_label = Label(master, text="Password Length:", cnf = lcnf_label1)
        self.length_label.grid(lcnf_grid_label1)
        
        self.length_entry = Entry(master, lcnf_entry)
        self.length_entry.grid(lcnf_grid_entry)

        self.generate_button = Button(master, text="Generate Password", command=self.generate_password)
        self.generate_button.config({'background': '#F08080',
                 'font':('Arial', 12, 'bold'), 
                 'foreground':'black',
                 'padx':10, 
                 'pady':10, 
                 'width':25, 
                 'height':1,
                 'relief':tk.RAISED})
        self.generate_button.grid(lcnf_generate_button)

        self.password_label = Label(master, text="Generated Password:", cnf=lcnf_label2)
        self.password_label.grid(lcnf_grid_label2)

        self.password_display = Entry(master, textvariable=self.password_var, cnf=lcnf_entry2)
        self.password_display.grid(lcnf_grid_entry2)
        
    def generate_password(self):
        try:
            length = int(self.length_entry.get())
            if length > 0:
                password = self.generate_random_password(length)
                self.password_var.set(password)
            else:
                self.password_var.set("Invalid length. Please enter a positive integer.")
        except ValueError:
            self.password_var.set("Invalid input. Please enter a valid integer.")
            
    def generate_random_password(self, length):
        
        uppercase = random.choice(string.ascii_uppercase)
        digit = random.choice(string.digits)
        special_char = random.choice(string.punctuation)

        # Generate the remaining characters
        remaining_length = length - 3

        characters = string.ascii_letters + string.digits + string.punctuation
        password = ''.join(random.choice(characters) for _ in range(remaining_length))

        # Shuffle and insert the required characters
        password_list = list(uppercase + digit + special_char + password)

        random.shuffle(password_list)

        return ''.join(password_list)
```

```python
def main():
    root = tk.Tk()
    app = PasswordGeneratorApp(root)
    root.mainloop()
```

```python
if __name__ == "__main__":
    main()
```
