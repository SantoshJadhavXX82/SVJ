```python
import tkinter as tk
from tkinter import messagebox
```

```python
lcnf1 = {'background':'black', 
         'borderwidth':2, 
         'font':('Arial', 30), 
         'foreground':'white', 
         'width':40}

lcnf2 = {'row':0, 'column':0, 'padx':10, 'pady':10}

lcnf_add_button = {'row':0, 'column':1, 'padx':10, 'pady':10}

lcnf_lstbx1 = {'background':'black', 
               'borderwidth':2, 
               'font':('Arial', 30), 
               'foreground':'white', 
               'width':40,
               'height':10}

lcnf_lstbx2 = {'row':1, 'column':0, 'padx':10, 'pady':10}

lcnf_view_button = {'row':2, 'column':0, 'padx':5, 'pady':10}

lcnf_complete_button = {'row':2, 'column':1, 'padx':10, 'pady':10}

lcnf_remove_button = {'row':2, 'column':2, 'padx':15, 'pady':10}
```

```python
class TodoListApp:
    def __init__(self, var1):
        self.var1 = var1
        self.var1.title('To do list Application')
        
        self.frame1 = tk.Frame(var1)
        self.frame1.grid(row=0, column=0, padx=10, pady=10)
        
        self.tasks = []
        
        self.task_entry = tk.Entry(self.frame1, lcnf1)
        self.task_entry.grid(lcnf2)
        
        self.add_button = tk.Button(self.frame1, text='Add Task', command=self.add_task)
        self.add_button.config({'background': '#F08080',
                 'font':('Arial', 12, 'bold'), 
                 'foreground':'black',
                 'padx':10, 
                 'pady':10, 
                 'width':12, 
                 'height':1,
                 'relief':tk.RAISED})
        self.add_button.grid(lcnf_add_button)
        
        self.task_listbox = tk.Listbox(self.frame1, lcnf_lstbx1)
        self.task_listbox.grid(lcnf_lstbx2)
        
        self.frame2 = tk.Frame(var1)
        self.frame2.grid(row=2, column=0, padx=0, pady=10)
        
        self.view_button = tk.Button(self.frame2, text="View Tasks", command=self.view_tasks)
        self.view_button.config({'background': '#F08080',
                 'font':('Arial', 12, 'bold'), 
                 'foreground':'black',
                 'padx':5, 
                 'pady':10, 
                 'width':15, 
                 'height':1,
                 'relief':tk.RAISED})
        self.view_button.grid(lcnf_view_button)

        self.complete_button = tk.Button(self.frame2, text="Mark as Completed", command=self.complete_task)
        self.complete_button.config({'background': '#F08080',
                 'font':('Arial', 12, 'bold'), 
                 'foreground':'black',
                 'padx':5, 
                 'pady':10, 
                 'width':15, 
                 'height':1,
                 'relief':tk.RAISED})
        self.complete_button.grid(lcnf_complete_button)

        self.remove_button = tk.Button(self.frame2, text="Remove Task", command=self.remove_task)
        self.remove_button.config({'background': '#F08080',
                 'font':('Arial', 12, 'bold'), 
                 'foreground':'black',
                 'padx':5, 
                 'pady':10, 
                 'width':15, 
                 'height':1,
                 'relief':tk.RAISED})
        self.remove_button.grid(lcnf_remove_button)
        
    def add_task(self):
        task = self.task_entry.get().strip()
        if task:
            self.tasks.append(task)
            self.task_listbox.insert(tk.END, task)
            self.task_entry.delete(0, tk.END)
        else:
            messagebox.showwarning("Warning", "Please enter a task.")

    def view_tasks(self):
        if not self.tasks:
            messagebox.showinfo("To-Do List", "No tasks in the list.")
        else:
            tasks_str = "\n".join(self.tasks)
            messagebox.showinfo("To-Do List", f"To-Do List:\n{tasks_str}")

    def complete_task(self):
        selected_index = self.task_listbox.curselection()
        if selected_index:
            task_index = selected_index[0]
            self.task_listbox.itemconfig(task_index, {'bg': 'light gray'})
            messagebox.showinfo("To-Do List", f'Task "{self.tasks[task_index]}" marked as completed.')
        else:
            messagebox.showwarning("Warning", "Please select a task.")

    def remove_task(self):
        selected_index = self.task_listbox.curselection()
        if selected_index:
            task_index = selected_index[0]
            removed_task = self.tasks.pop(task_index)
            self.task_listbox.delete(task_index)
            messagebox.showinfo("To-Do List", f'Task "{removed_task}" removed from the list.')
        else:
            messagebox.showwarning("Warning", "Please select a task.")
```

```python
def main():
    root = tk.Tk()
    app = TodoListApp(root)
    root.mainloop()
```

```python
if __name__ == "__main__":
    main()
```
