```python
download_url = ("https://raw.githubusercontent.com/fivethirtyeight/"
                "data/master/college-majors/recent-grads.csv")
```

```python
df = pd.read_csv(download_url)
```

```python
pd.set_option('display.max.columns', None)
df
```

```python
#%matplotlib
```

```python
%matplotlib inline
```

```python
df.plot(x='Rank', y=['P25th', 'Median', 'P75th'])
plt.show()
```

```python
df.plot(kind='line',x='Rank', y='P75th')
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df['Rank'], df['P75th'])
```

```python
df.plot(x='Rank', y='P75th')
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(8,4))

ax1.plot('Rank','P75th', data=df)
ax1.legend(loc=(0))
ax1.set_title('Line: $x$ versus $y$')
ax1.grid(linestyle='--')
ax1.set_axisbelow(True)

ax2.scatter(data=df,x='Rank', y='P25th', marker='o')
ax2.legend(loc=(0))
ax2.set_title('Frequencies of $x$ and $y$')
ax2.yaxis.tick_right()
ax2.grid(linestyle='--')
ax2.set_axisbelow(True)

plt.show()
```
